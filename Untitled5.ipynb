```python
#Identifying Outliers with Interquartile Range (IQR)
#statistical dispersion and is calculated as the difference between the 75th and 25th percentiles. 
#It is represented by the formula IQR = Q3 − Q1. A quantile defines a particular part of a data set.
#quantile determines how many values in a distribution are above or below a certain limit. Special quantiles are the 
#quartile (quarter).f we divide a distribution into four equal portions, we will speak of four quartiles. 
#The first quartile includes all values that are smaller than a quarter of all values. In a graphical representation, it 
#corresponds to 25% of the total area of a distribution. The two lower quartiles comprise 50% of all distribution values. The 
#interquartile range between the first and third quartile equals the range in which 50% of all values lie that are distributed around the mean.
from scipy.stats import iqr
#print(train_df.revenue)
Q1= train_df.revenue.quantile(0.25)
Q3=train_df.revenue.quantile(0.75)

IQR=Q3-Q1
 #or do as IQR= iqr(train_df.revenue)
#upper_limit = train_df.revenue.quantile(0.75) + (1.5* iqr(train_df.revenue))
#lower_limit = train_df.revenue.quantile(0.25)- (1.5* iqr(train_df.revenue))
```

```python

#condition = (train_df.revenue > upper_limit) | (train_df.revenue<lower_limit)
#or
condition=(train_df.revenue < (Q1 - 1.5 * IQR)) |(train_df.revenue > (Q3 + 1.5 * IQR))
train_df[condition]

```

```python
#Combine training set and test set into complete set for formatting.
comp_df = pd.concat([train_df, test_df])
comp_df.reset_index(drop=True, inplace=True)
```

```python
comp_df.Type.value_counts(),comp_df['City Group'].value_counts()
```

```python
comp_df.Type = comp_df.Type.map({'MB':0,'DT':1, 'IL':2,'FC':3})
comp_df['City Group'] = comp_df['City Group'].map({'Big Cities':1, 'Other':0})
```

```python
comp_df.head(3)
```

```python
#Normalize the p-columns
p_name = ['P'+str(i) for i in range(1,38)]
comp_df[p_name] = MinMaxScaler().fit_transform(comp_df[p_name])
#comp_df[p_name].head
```

```python
from sklearn.decomposition import PCA
pca=PCA().fit(comp_df[p_name])
plt.figure(figsize=(7,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('Expalined Variance')
plt.yticks(np.arange(0.1,1.1,0.05))
plt.xticks(np.arange(0,41,2))
plt.show(True)
```

```python
pca_list = ['pca'+str(i) for i in range(1,30,1)]
#print(pca_list)
#add multiple pca columns to the comp_df dataframe
comp_df=pd.concat([comp_df,pd.DataFrame(columns=pca_list)])
#comp_df.info()
comp_df[pca_list] = PCA(n_components=29).fit_transform(comp_df[p_name])
comp_df.drop(p_name,axis=1,inplace=True)
```

```python
#Normalize the date
#Create new column 'launch days'
#Normalize open_year, launch days
import datetime
comp_df['launch_days'] =  (datetime.datetime.now() - comp_df[['Open Date']])
comp_df['launch_days'] = comp_df['launch_days'].dt.days
```

```python
comp_df['launch_days'] = MinMaxScaler().fit_transform(comp_df[['launch_days']])
comp_df['open_year'] = MinMaxScaler().fit_transform(comp_df[['open_year']])
```
